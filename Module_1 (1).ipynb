```python
#Dropping irrelevant columns
data.drop(['ID','Name','Location','Family_Communication'], inplace = True, axis=1)
```

```python
data.info()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(data[['Daily_Usage_Hours','Sleep_Hours','Academic_Performance','Time_on_Social_Media'
,'Anxiety_Level','Depression_Level','Self_Esteem','Time_on_Gaming','Addiction_Level','Time_on_Education']].corr(),annot=True,  cbar=True)

```

```python
#used chatgpt assistance to determine how to evenly divide values into 3 categories
data['Academic_Performance_Category'] = pd.cut(data['Academic_Performance'], bins=3, labels =['Low','Medium','High'])
```

```python
sns.set_palette("pastel")
plt.title('Screen Time Before Bed vs Academic Performance')
sns.barplot(data,y='Screen_Time_Before_Bed',x='Academic_Performance_Category')
```

```python
plt.title('Anxiety Level vs Time On Social Media')
sns.boxplot(data, y='Time_on_Social_Media', x='Anxiety_Level')
```

```python
plt.title('Anxiety Level vs Time On Education')
sns.boxplot(data, y='Time_on_Education', x='Anxiety_Level')
```

```python
plt.title('Anxiety Level vs Time On Gaming')
sns.boxplot(data, y='Time_on_Gaming', x='Anxiety_Level')
```
